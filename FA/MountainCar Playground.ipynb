```python
%matplotlib inline

import gym
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Create the environment with render_mode='rgb_array'
env = gym.make("MountainCar-v0", render_mode='rgb_array')

# Reset the environment
env.reset()

# Take 10,000 steps in the environment
for _ in range(10000):
    env.step(0)  # Perform action 0 (do nothing)

# Render the environment and display the frame
frame = env.render()  # No need to pass 'mode' argument
plt.figure()
plt.imshow(frame)
plt.show()

# Close the environment
env.close()
```

```python
import gym
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Create environment with RGB array rendering
env = gym.make("MountainCar-v0", render_mode='rgb_array')

# Reset environment
obs, _ = env.reset()

# Collect frames for animation
frames = []

for _ in range(100):  # Adjust number of frames as needed
    obs, reward, done, truncated, info = env.step(0)
    frame = env.render()
    frames.append(frame)
    if done or truncated:
        break

env.close()

# Create animation using matplotlib
fig, ax = plt.subplots()
img = ax.imshow(frames[0])
ax.axis('off')  # Optional: remove axis for clean display

def update(i):
    img.set_data(frames[i])
    return [img]

ani = animation.FuncAnimation(
    fig, update, frames=len(frames), interval=500, blit=True
)

plt.title("MountainCar-v0 Animation")
plt.show()
```

```python
ani.save("mountaincar_animation.gif", writer='pillow', fps=20)
```
