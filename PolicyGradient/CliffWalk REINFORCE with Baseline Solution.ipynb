# REINFORCE with Baseline

```python
%matplotlib inline

import matplotlib
import sys

if "../" not in sys.path:
  sys.path.append("../") 
from lib.envs.cliff_walking import CliffWalkingEnv

from REINFORCE import ValueEstimator, PolicyEstimator, reinforce
matplotlib.style.use('ggplot')
```

![](../figs/REINFORCE.png)

```python
env = CliffWalkingEnv()
num_states = env.observation_space.n
num_actions = env.action_space.n

# Initialize estimators
policy_estimator = PolicyEstimator(num_states, num_actions, learning_rate=1e-2)
value_estimator = ValueEstimator(num_states, learning_rate=1e-2)

# Train using REINFORCE
stats = reinforce(env, policy_estimator, value_estimator, num_episodes=5000, discount_factor=1)
```

```python
print(stats.episode_lengths[-1])
print(stats.episode_rewards[-1])
```

```python
from lib import plotting
plotting.plot_episode_stats(stats, smoothing_window=25)
```
